# Q2. Implement Insertion sort algorithm.

```python
import time
```

```python
def InsertionSort(Array):
    for j in range(1,len(Array)):
       key = Array[j]
       i = j-1
       while i>=0 and Array[i]>key:
           Array[i+1] = Array[i]
           i = i-1
       Array[i+1] = key
    return Array;
```
